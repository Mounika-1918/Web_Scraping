```python
import requests
from bs4 import BeautifulSoup
page=requests.get("https://www.geeksforgeeks.org/top-indian-ceos-in-the-world/")
soup=BeautifulSoup(page.text)
title=soup.find('h1')
print(title.text)
```

```python
import pandas as pd
df=pd.DataFrame(title)
df
```

```python
table=soup.find('div',id="table_of_content")
print(table)
```

```python

a_tags=table.find_all('a')
for a_tag in a_tags:
    print(a_tag.text)
```
